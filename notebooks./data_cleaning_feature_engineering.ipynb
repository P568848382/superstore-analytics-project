Advance feature engeneering

A)Shipping efficiency

In buisness"Order to Ship" time is a key KPI

```python
#we need to create some new columns that provides buisness values.
#shipping efficiency
orders['shipping_days']=(orders['ship_date']-orders['order_date']).dt.days
#to check are there any negative shipping days?(data errors)

orders[orders['shipping_days']<0]
#as the result showing there is no such date where order_date is more than ship_date in simple words there are 
# no such type or orders who has shipped before ordered :P
```

B)Profitability Analysis
    

```python
# Profitability Analysis
#Unit Price
orders['unit_price']=orders['sales']/orders['quantity']
# Profit Margin 
orders['profit_margin']=(orders['profit']/orders['sales']).round(2)
```

Time Based Features 

A)Yearly Growth:-are we making more money year by year

B)Monthly Seasonlity:-Do sales spikes on particular month.On which months we need to more marketing campaigns.

C)Day of Week:-on which day our sales are increasing .On particular day if sales spikes then we need extra staff in the warehouse to process them



```python
orders['order_year']=orders['order_date'].dt.year
orders['order_month']=orders['order_date'].dt.month
orders['order_month_name']=orders['order_date'].dt.month_name()
orders['order_day_of_week']=orders['order_date'].dt.day_name()
```

```python
#there is no need of row_id so we drop that column as well
orders.drop(columns='row_id',inplace=True)
```

Applying qcut to shipping days to find relative speed of our logistics.It helps management identify which shipping segement is failing.

```python
# We want 4 categories: 'Critical', 'High', 'Medium', 'Low'
# Note: In shipping, 0 days is 'Critical' (Fastest), so we label accordingly.
orders['shipping_efficiency']=pd.qcut(orders['shipping_days'],q=4,labels=['Critical','High','Medium','Low'])
print(orders['shipping_efficiency'].value_counts())
```

The Risks of ignoring this:

Duplicate Records: When we try to create a dim_products table in SQL, we'll get 4 rows for the same ID. This breaks "Referential Integrity."
Our product_ids showing one product_id has 4 different product_names.

How to fix it.We do Data Normalization:
We need to ensure each product_id has only one "Official Name."



```python
#create a unique product_id based on Name + existing Id
orders['unique_product_id']=orders['product_id'] + '_' + orders['product_name']
```

Now move this data frame inside the sql 
we shall use pgsql

```python
from sqlalchemy import create_engine
```

```python
#Prepare tables(star Schema)
dim_customers=orders[['customer_id','customer_name','segment']].drop_duplicates()
dim_products=orders[['unique_product_id','product_name','category','sub_category']].drop_duplicates()
#our dim_location city,state,and country but our fact_orders table does not have a single 'ID' to link them.We need to create 'location_id'
#create a unique list of locations.
dim_locations=orders[['city','state','country','market','region','postal_code']].drop_duplicates().reset_index(drop=True)
#add a unique ID column (eg.LOC_1,LOC_2...)
dim_locations['location_id']=dim_locations.index + 1
#  Prepare the orders dataframe for the Fact Table
# We must merge the location_id back so the Fact Table has a reference to the locations
orders = orders.merge(dim_locations, on=['city', 'state', 'country', 'market', 'region', 'postal_code'], how='left')

#  Create fact_orders
# We keep IDs for linking and all numerical/metric columns
fact_orders = orders[[
    'order_id', 'order_date', 'ship_date', 'ship_mode', 'customer_id', 
    'unique_product_id', 'location_id', 'sales', 'quantity', 'discount', 
    'profit', 'shipping_cost', 'order_priority', 'shipping_days', 
    'unit_price', 'profit_margin', 'order_year', 'order_month', 
    'order_day_of_week', 'shipping_efficiency'
]]
```

```python
engine=create_engine('postgresql://postgres:1234@localhost:5432/superstore_db')
```

```python
# # Push tables to PostgreSQL
# # 'if_exists=replace' ensures that if we re-run the script, it updates the tables
dim_customers.to_sql('dim_customers', engine, if_exists='replace', index=False)
dim_products.to_sql('dim_products', engine, if_exists='replace', index=False)
dim_locations.to_sql('dim_locations', engine, if_exists='replace', index=False)
fact_orders.to_sql('fact_orders', engine, if_exists='replace', index=False)
print("Star Schema successfully pushed to PostgreSQL!")
```
